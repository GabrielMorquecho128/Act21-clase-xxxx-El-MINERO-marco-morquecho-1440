```python
# Act21 clase Minero 39 [Tu Nombre] 1777

class Minero:
    def __init__(self, nombre, telefono, edad):
        self.nombre = nombre    # Atributo 1
        self.telefono = telefono # Atributo 2
        self.edad = edad        # Atributo 3

    def barrenar(self): return f"⚡ {self.nombre} está BARRENANDO."
    def picar(self): return f"⛏️ {self.nombre} está PICANDO."
    def explorar(self): return f"🔦 {self.nombre} está EXPLORANDO."

# --- Creación de Objetos ---
m1 = Minero("JUAN-01", "555-101", 34)
m2 = Minero("PEDRO-02", "555-202", 29)

# --- Resultados ---
for m in [m1, m2]:
    print(f"\nOBJETO: {m.nombre} | Tel: {m.telefono} | Edad: {m.edad}")
    print(m.barrenar())
    print(m.picar())
    print(m.explorar())

print("programa realizado por marco morquecho NL = 39")
```
